2 * pi / M * i [0,1,2,3,4,5,6,7]

```python
phase = 2 * np.pi / M * 0

ic(np.rad2deg(phase))

phase_amp = np.sqrt(1)

ic(phase_amp)

base_symbol = phase_amp * np.exp(1j * phase)

ic(base_symbol)
```

Also erstellt man sich ein array mit einer forschleife das quasi alle winkelpositionen h채lt

```python
phase_angles = np.array([2 * np.pi / M * np.arange(M)])

ic(np.rad2deg(phase_angles))

phase_symbols = phase_amp * np.exp(1j * phase_angles)

ic(phase_symbols)

```
